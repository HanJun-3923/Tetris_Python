<h4>01. 프로젝트의 주제<br></h4>
본 Term Project의 주제는 '유전 알고리즘을 이용한 인공지능 테트리스'이다. 유전 알고리즘을 사용해 인공지능을 만든 뒤, 인공지능과 테트리스로 직접 대결할 수 있는 게임을 만드는 것이다. 인공지능은 블럭을 놓는 방법에 대한 여러가지 규칙들을 지니며, 각 규칙은 적절한 가중치 값을 가진다. 규칙과 가중치를 통해 최적의 위치를 찾아낸 뒤, 해당 위치에 블럭을 놓는 것이 인공지능의 가장 기본적인 작동 원리이다. 또한, 가중치의 값을 수동으로 조절하여 최적의 가중치 값을 얻는 것은 효율적이지 않기에 '유전알고리즘'을 통해 가중치 값을 얻는다.<br>
<br>
이를 통해 인공지능 관점의 사고 능력을 키우고 인공지능의 기본적인 작동 원리와 유전알고리즘에 대한 이해도 및 사용법을 배울 수 있다. 또한, 인공지능 테트리스를 통해 효율적인 테트리스에 대해 공부하여 사고력을 확장하는 것은 물론, 인공지능의 수준을 고려해 적절한 연습 대상을 얻을 수도 있다.


<h4>02. 유전 알고리즘</h4>

유전 알고리즘(Genetic Algorithm)은 자연세계에서 생물의 진화 과정을 모방하여 만든 계산 모델이다. 본 프로젝트에서는 해당 위치에서 블럭의 적합도를 계산하기 위한 규칙들의 적절한 가중치 값을 구하는데에 사용한다. 유전 알고리즘을 사용하기 위해선 몇 가지 고려해야 할 조건들이 있다.<br>

1. 해를 유전자 형식으로 표현할 수 있어야 한다.
2. 해가 얼마나 적합한지 적합도 함수를 통해 계산하여 수치로 표현할 수 있어야 한다.
3. 유전 알고리즘의 연산(선택, 교차, 변이, 대치)을 적용할 수 있어야 한다.<br>

규칙들의 가중치들이 수치화된 유전자이며, 게임을 플레이하여 얻은 APM(Attack Per Minute), PPS(Piece Per Second), Score 등의 수치를 통해 적합도 함수를 제작할 수 있기에 유전 알고리즘은 인공지능 테트리스를 만들기에 적합하다.



<h4>03. 테트리스의 규칙</h4>

(1) 테트로미노<br>
테트리스는 정사각형 블럭 4개를 이어 붙여 만든 7개의 모양이 존재하며 이를 '테트로미노'라고 부른다. 추가적인 테트로미노는 이후 적용할 규칙들에 적합하지 않으므로 표준적으로 사용되는 7개의 테트로미노만 사용한다.<br>

(2) 테트리스의 목적<br>
테트리스는 크게 2가지 방식이 존재한다. 오래 살아남아 고득점을 얻는 방식과 상대방과 대전하여 겨루는 방식이 있는데, 본 프로젝트의 테트리스는 상대방과 대전하여 겨루는 방식의 테트리스이다. 따라서 오래 살아남으면서 상대방에게 최대한 많은 피해를 주는 것이 인공지능의 이상적인 모습이다.<br>

</h5>테트리스의 기능<br>
(3-1) Super Rotation System(SRS): 블럭이 장애물과 회전하는 과정에서 충돌이 발생할 때 이를 처리하는 오프셋과 관련있는 규칙이다. T-Spin을 포함한 다양한 Spin과 보편적인 피네스(Finesse)와 Kick 기능이 SRS를 통해 구현된다. (피네스란, 블럭을 원하는 위치에 놓기 위한 최소한의 블럭 이동 방식을 의미한다.)<br>

(3-2) Attack Table: 필드에서 한 행을 모두 채워 라인을 지웠을 때, 상대방에게 얼마나 많은 공격을 보낼지에 대한 기준이다. 각각의 테트리스마다 조금씩 다르기에 본 프로젝트의 테트리스는 BTB을 중요시하는 방식의 Attack Table을 제작하여 사용한다. (이때 BTB이란, 4줄을 한 번에 지우는 '테트리스'와 'T-Spin'을 몇 번 연속적으로 하였는지 보여주는 수치이다.) 따라서 사용자가 속도보다 효율성을 더욱 중시하여 플레이 하도록 유도한다.<br>

(3-3) Ghost<br>
실시간으로 하드드롭을 눌렀을 때 블럭이 놓일 위치에 고스트 블럭을 보여주어 사용자에게 블럭이 놓일 위치에 대한 가이드라인을 제공해준다.<br>
